```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('iris.csv')
x = data['SepalLengthCm']
y = data['SepalWidthCm']
z = data['PetalLengthCm']
```

```python
plt.scatter(x,y)
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("SCATTER PLOT")
plt.show()
```

```python
box_data = data.drop(columns=['PetalLengthCm','PetalWidthCm','Id'])
sns.boxplot(data=box_data)
plt.title('BOX PLOT')
plt.show()
```

```python
heat_data = pd.get_dummies(data,columns=['Species'])
sns.heatmap(heat_data.corr(),annot=True,cmap="jet_r")
plt.title("HEAT MAP")
plt.show()
```

```python
plt.tricontour(x,y,z,cmap="jet")
plt.title("CONTOUR PLOT")
plt.show()
```

```python
ax = plt.axes(projection = "3d")
ax.plot_trisurf(x,y,z,cmap="jet_r")
ax.set_title("3D SURFACE PLOT")
plt.show()
```
