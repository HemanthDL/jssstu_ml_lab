```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier as KNN
from sklearn.metrics import accuracy_score

glass = pd.read_csv('glass.csv')
x = glass.drop(columns=['Type'],axis=1).values
y = glass['Type'].values
fruit = pd.read_csv('fruits.csv')
xf = fruit[['mass','width','height','color_score']].values
yf = fruit['fruit_label']

datasets = ['glass','fruit']

K = [3,5,7]
distancemetircs = ['manhattan','euclidean']

splitdata = [0.1,0.3]

for d in datasets:
    print(f"...For Dataset - {d}... ")
    for k in K:
        print("\tK = ",k)
        for p,metrics in enumerate(distancemetircs):
            print(f"\t\tDistance Metric : {metrics}")
            for ts in splitdata:
                print(f"\t\t\ttest_split : {ts}")
                knn = KNN(n_neighbors = k,p=p+1,weights="distance")
                if(d == 'glass'):
                    xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=ts,random_state=42)
                else:
                    xtrain,xtest,ytrain,ytest = train_test_split(xf,yf,test_size=ts,random_state=42)
                knn.fit(xtrain,ytrain)
                ypred = knn.predict(xtest)
                acc = accuracy_score(ypred,ytest)
                print(f"\t\t\tAccuracy : {acc : .4f}")
                print("\n")


```
