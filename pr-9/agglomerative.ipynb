```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram,linkage
from sklearn.datasets import load_iris

data = load_iris().data

data = data[:6]

def proximity(data):
    n = data.shape[0]
    p = np.zeros((n,n))
    for i in range(n):
        for j in range(i+1,n):
            p[i,j] = np.linalg.norm(data[i]-data[j])
            p[j,i] = p[i,j]

    return p

def plotdendro(data,method):
    mat = linkage(data,method=method)
    print(f"\n\nCluster Matrix :\n\n{mat}\n\n ")
    dendrogram(mat)
    plt.title(f"{method} Linkage")
    plt.show()

print(f"Proximity Matrix : \n ",proximity(data))

plotdendro(data,'single')
plotdendro(data,'complete')
```
