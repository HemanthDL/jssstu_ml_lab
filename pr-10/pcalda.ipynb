```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA

x = load_iris().data
y = load_iris().target
print(f"\n\n Original Data : \n\n{x[:6]}\n\n")
print("Shape of Original Dataset: ",x.shape)

plt.scatter(x[:,0],x[:,1],c=y,cmap="jet")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Original")
plt.show()

pca = PCA(n_components=2)
pca.fit(x)
new_x = pca.transform(x)
print("PCA..\n")
print(f"Transformed Data : \n\n{new_x[:6]}\n\n")
print("Shape of Transformed Data : ",new_x.shape)

plt.scatter(new_x[:,0],new_x[:,1],c=y,cmap="jet")
plt.xlabel("Principle Component 1")
plt.ylabel("Principle Component 2")
plt.title("PCA")
plt.show()

lda = LDA(n_components=2)
lda.fit(x,y)
new = lda.transform(x)
print("\n\nLDA...\n")
print(f"Transformed Data : \n\n{new[:6]}\n\n")
print("Shape of Transformed Data : ",new.shape)

plt.scatter(new[:,0],new[:,1],c=y,cmap="jet")
plt.xlabel("Linear Discriminant 1")
plt.ylabel("Linear Discriminant 2")
plt.title("LDA")
plt.show()
```
